```python
%matplotlib inline

import os
import sys
parent_path = os.path.abspath(os.path.join('..'))
if parent_path not in sys.path:
    sys.path.append(parent_path)

import numpy as np
import pandas as pd
import time

from preprocess.data_handler import load_data
from preprocess.rdkutils import fp_from_smiles
from preprocess.smiles_embedder import get_cnn_fingerprint

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_samples, silhouette_score

from ggplot import *
```

Selected 10 terms (the most frequent ones), data are selected only if not associated with more than one term in the list.

```python
sel_terms = ['Anti-Bacterial Agents', 'Enzyme Inhibitors', 'Antineoplastic Agents', 'Antihypertensive Agents',
             'Vasodilator Agents', 'Anti-Inflammatory Agents, Non-Steroidal', 'Anti-Arrhythmia Agents',
             'Antiviral Agents', 'Antioxidants', 'Anticonvulsants']
path = '../data/'
df = load_data(path+'dataset.csv')
sel_data = pd.DataFrame([])

for t in sel_terms:
    check = [i for i in sel_terms if i != t]
    l = []
    for i, terms in enumerate(df['Terms']):
        if t in terms and t not in check:
            if len(list(set(check)-set(terms))) == len(check):
                l.append([df.at[i, 'CID'], df.at[i, 'SMILES'], t])
    d = pd.DataFrame(l, columns=['CID', 'SMILES', 'Term'])
    sel_data = sel_data.append(d, ignore_index=True)

print('Selected examples: ', sel_data.shape[0])
```

Embedding smiles using CNN-FP and ECFP

```python
smiles = sel_data['SMILES']
X_ecfp, _ = fp_from_smiles(smiles, 2, 512, 'ecfp')
X = get_cnn_fingerprint(smiles)
y = sel_data['Term']
```

Before applying tSNE, it's reccomended to reduce the number of dimensions using, for example, PCA

```python
pca = PCA(n_components=100)
pca_result = pca.fit_transform(X)

pca_ecfp = PCA(n_components=100)
pca_result_ecfp = pca_ecfp.fit_transform(X_ecfp)

print('Explained variation per principal component (PCA) - CNN-FP: {}'.format(np.sum(pca.explained_variance_ratio_)))
print('Explained variation per principal component (PCA) - ECFP: {}'.format(np.sum(pca_ecfp.explained_variance_ratio_)))
```

Transform data into tSNE space

```python
time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(pca_result)
print('CNN-FP | t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))

time_start = time.time()
tsne_ecfp = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results_ecfp = tsne_ecfp.fit_transform(pca_result_ecfp)
print('ECFP | t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

Visualize the two resulting dimensions in a scatter plot, coloring each sample by the respective label

CNN-FP

```python
df_tsne = pd.DataFrame([])
df_tsne['x-tsne'] = tsne_results[:, 0]
df_tsne['y-tsne'] = tsne_results[:, 1]
df_tsne['label'] = y

chart = ggplot(df_tsne, aes(x='x-tsne', y='y-tsne', color='label')) \
        + geom_point(size=70, alpha=0.1) \
        + ggtitle("tSNE dimensions colored by term")
chart
```

```python
print("Silhouette Coefficient: %0.3f"
      % silhouette_score(tsne_results, y))

print("Silhouette Coefficient per each sample:", silhouette_samples(tsne_results, y))
```

ECFP

```python
df_tsne_ecfp = pd.DataFrame([])
df_tsne_ecfp['x-tsne'] = tsne_results_ecfp[:, 0]
df_tsne_ecfp['y-tsne'] = tsne_results_ecfp[:, 1]
df_tsne_ecfp['label'] = y

chart_e = ggplot(df_tsne_ecfp, aes(x='x-tsne', y='y-tsne', color='label')) \
        + geom_point(size=70, alpha=0.1) \
        + ggtitle("tSNE dimensions colored by term - ECFP")
chart_e
```

```python
print("Silhouette Coefficient: %0.3f"
      % silhouette_score(tsne_results_ecfp, y))

print("Silhouette Coefficient per each sample:", silhouette_samples(tsne_results_ecfp, y))
```
